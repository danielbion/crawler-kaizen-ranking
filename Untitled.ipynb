```python
import time
from tqdm import tqdm
data = pd.DataFrame({'Rank': [], 'Name': [], 'Guild': [], 'Job': [], 'Level': []})

for i in tqdm(range(0,100)):
    
    if i%10 == 0:       
        view_page = session.get("https://kaizenms.net/?base=main&page=rankings&job=overall&start=" + str(i), headers=headers)

        soup = BeautifulSoup(view_page.content, 'html.parser')
        rows = soup.find_all('tr')

        for row in rows[1:11]:
            cols = row.find_all("td")
            rank = cols[0].get_text()
            name_guild = cols[2].get_text().strip().split('\n')
            name = name_guild[0]
            guild = ''
            if len(name_guild) > 1:
                guild = name_guild[1]
            job = cols[3].find('img').next_sibling.next_sibling.strip()
            level = cols[3].find("td").find('b').get_text()
            data = data.append({'Rank': rank, 'Name': name, 'Guild': guild, 'Job': job, 'Level': level}, ignore_index = True)
data.to_csv('ranking.csv', index = False)
```

```python
def get_group_by_level(level_min, level_max):
    g = data[(data.Level >= level_min) & (data.Level < level_max)]\
    .groupby('Job', as_index = False).agg({'Level': 'mean', 'Rank': 'count'})\
    .sort_values('Rank', ascending = False)\
    .rename(columns={'Level': 'Media de level', 'Rank': 'Players'})
    g['Media de level'] = g['Media de level'].map(round)
    return g
```
